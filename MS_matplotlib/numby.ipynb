```python
import numpy as np 
import matplotlib. pyplot as plt
```

```python
# Define the function
def f(x):
    return (2*x*x) + 4
# Generate x values
x_values = np.linspace (-2, 3, 50)
# Calculate y values
y_values = f(x_values)
# PLot the function
plt.plot(x_values, y_values, label= 'f(x) = 2x^2 +4')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x - 2')
plt.grid(True)
plt.legend()
plt.show()
```
