```python
import matplotlib.pyplot as plt

#du lieu mau
height = [176, 187, 152, 170, 195, 158, 175, 169, 190, 155]
weight = [80, 98, 45, 76,112, 65, 78, 70, 105, 50]

#ve bieu do phan tan
plt.scatter(height, weight, color='blue', marker='+', s=50)

#thiet lap nhan cho truc x va truc y
plt.xlabel('height')
plt.ylabel('weight')

#dat tieu de
plt.title('height-weight plot')

#bat luoi cho bieu do
plt.grid(True)

#hien thi bieu do
plt.show()
```

```python
import matplotlib.pyplot as plt

#du lieu mau
club = ['Chelsea', 'Manchester City', 'Manchester United', 'Arsenal', 'Tottenham Hotspur', 'Others']
supporter = [35, 25, 15, 10, 10, 5]

#ve bieu do duong
plt.pie(supporter, labels=club, shadow=True, autopct='%1.1f%%', startangle=180)

#them legend
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))

#thiet lap truc tam
plt.axis('equal')

#thiet lap nhan cho truc x va truc y
# plt.xlabel('months')
# plt.ylabel('temps')

#dat tieu de
plt.title('football supporters as to who will win the Premier League in 2019/20')

#bat luoi cho bieu do
# plt.grid(True)

#hien thi bieu do
plt.show()
```

```python
from collections import Counter
```

```python
children = [4, 5, 2, 2, 3, 4, 4, 3, 5, 4, 
7, 3, 3, 4, 2, 2, 2, 2, 2, 6, 
3, 2, 3, 3, 1, 2, 3, 2, 2, 6, 
5, 5, 3, 2, 4, 4, 2, 4, 1, 2, 
2, 2, 1, 3, 3, 2, 2, 4, 5, 3]
```

```python
f_children = dict(Counter(children))
f_children
```

```python
#ve bieu do cot
plt.bar(list(f_children.keys()), list(f_children.values()), color='skyblue')

#them tieu de va nhan cho truc
plt.xlabel('Number of children')
plt.ylabel('frequency')
plt.title('a bar chart')

plt.show()
```
